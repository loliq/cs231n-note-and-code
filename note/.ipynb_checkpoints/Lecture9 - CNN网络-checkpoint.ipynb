### 3. GoogleNet
特点：
1. 22 层
2. inception Model
3. 没有全连接层，节省了大量参数
4. 只有5 million的参数
5. 使用了bottleNeck层(1x1卷积)提高效率

### 4. ResNet
- 152层的网络
- 在很多benchmark 上表现都是最好
- 没有全连接层

直觉是：与其直接学习$H(x)$,不若学习$H(x)$与$x$的差异，也就是$F(x)$,$F(x)$也叫做残差，这样会更简单。

为什么学习残差会比直接学习影射$H(x)$更简单，这是一种假设。假设当恒等映射是好的，则将$F(x)$压缩至0即可，这样似乎使学习更简单。
